```python
import requests
from bs4 import BeautifulSoup
from zipfile import ZipFile, ZIP_DEFLATED
import os
import glob
from pypdf import PdfReader

# Упаковка в zip-архив
def to_zip():
    file_dir = glob.glob('./pdf/*')
    with ZipFile('./pdf.zip', 'w', compression=ZIP_DEFLATED) as zpf:
        for pdf_file in file_dir:
            zpf.write(pdf_file)

# Извлечение текста из pdf
def text_extractor(to_path):
    file_dir = glob.glob('./pdf/*.pdf')
    reader = PdfReader(to_path + '0.pdf')
    number_of_pages = len(reader.pages)
    
    for elem in file_dir: 
        file1=open('./pdf/' + elem[5:-4] + '.txt',"a")    
    
        for i in range(number_of_pages):
            page = reader.pages[i]
            file1.writelines(page.extract_text())   
    
# Загрузка pdf с сайта   
def get_pdf(url:str, path: str):

    if not os.path.isdir(path):
        os.makedirs(path)
    
    request = requests.get(url)
    soup = BeautifulSoup(request.content, 'lxml') 
    
    pdf = soup.find_all('a', href=True) 
    
    for item in pdf: 
        href = item['href']        
        if(href[-4:]=='.pdf'): 
            pstu = 'https://pstu.ru/' + href  
             
            # Выводим первые 10 pdf
            for i in range(10):   
                response = requests.get(pstu)               
                out = open(path + str(i) + '.pdf', 'wb')
                out.write(response.content)
                out.close()  
            break 
    
           
if __name__ == "__main__":  
            
    get_pdf(    
        url = 'https://pstu.ru/sveden/education/',            
        path = '/home/kirill/projects/Web_Data_Mining/Web-Data-Mining/Practics/practic_4/pdf/'
    )
    
    text_extractor(
        '/home/kirill/projects/Web_Data_Mining/Web-Data-Mining/Practics/practic_4/pdf/'
    )
    
    to_zip()
```
